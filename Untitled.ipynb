```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('FrenchRetail.csv', sep=';')
y = df['French_Retail_Sales']
```

```python
class model:
    x = df['French_Real_GDP']
    
    def __init__(self, data, quarters=False, month=False, last=False, MSFE = False, GDP = False, h=0, april=False):
        ''' Quarters OR month'''
        self.data = data
        self.quarters = quarters
        self.month = month
        self.last = last 
        self.MSFE = MSFE
        self.GDP = GDP
        self.h = h
        self.x = self.x.shift(periods=h)[h:]
        self.y = df['French_Retail_Sales']
        self.y = self.y[h:]
        self.april = april
    def __repr__(self):

        return 'Quarters dummy: {}\nMonths dummy: {}\nOnly Last: {}\nParameters: {}\nModel Evals: {}'.format(
            self.quarters, self.month, self.last, self.parameter_estimation(), self.model_eval())
        #Print model specification, parameter estimation, model_eval

    def Specification(self):
        '''Changes the dataframe to be in quarter or month form'''
        df = pd.read_csv(self.data, sep = ';')
        df['DATE'] = pd.to_datetime(df['DATE'], dayfirst = True)
        if self.quarters:
            df['DATE']=df['DATE'].dt.quarter
            return df
        if self.month:
            df['DATE']=df['DATE'].dt.month
            return df
    
    def dummies(self):
        '''Makes dummies based on if quarter or month'''
        mod = self.Specification()
        X = pd.get_dummies(mod['DATE'])
        X = X.reset_index()
        X.rename(columns={'index':'t'}, inplace=True)
            #Also needs an intercept
        if self.last:
            X.insert(0,'a0',1)
            if self.GDP:
                X = X.join(self.x)
                X = X[self.h:]
                return X.iloc[:,:2].join(X.iloc[:,-2:])
            else:
                return X.iloc[:,:2].join(X.iloc[:,-1:])
        if self.GDP:
            X = X.join(self.x)
            X = X[self.h:]
        return X
    
    def parameter_estimation(self):
        '''Estimates the parameters for the model'''
        X = self.dummies()
        if self.GDP and self.last:
            #print(X)
            #print(self.y)
            a0, t, dummy, gdp = np.linalg.inv((X.T@X)) @ X.T@self.y
            return a0, t, dummy, gdp
        if self.last:
            a0, t, dummy = np.linalg.inv((X.T@X)) @ X.T@self.y
            return a0, t, dummy
        if self.GDP:
            #print(np.linalg.inv((X.T@X)) @ X.T@self.y)
            t, gdp, *dummy = np.linalg.inv((X.T@X)) @ X.T@self.y
            return t, gdp, dummy
        else:
            t, *dummy = np.linalg.inv((X.T@X)) @ X.T@self.y
            return np.append(t, dummy)
    
    def y_hat(self):
        matrix = np.array(self.parameter_estimation())
        return self.dummies()@matrix.T

    def model_eval(self):
        MSE = np.mean((self.y - self.y_hat())**2)
        T = len(self.y)
        AIC = T * MSE + (5*2)
        BIC = T * MSE + 5*np.log(T)
        return MSE, AIC, BIC
    
    def MSFE_1(self, T0, h, plot=False):
        X = self.dummies()
        T = len(self.y)
        syhat = []
        for i in range(T0, T-h):
            temp_y = self.y[:i]
            temp_X = X[:i]
            temp_X = np.array(temp_X)
            beta_hat = (np.linalg.inv((temp_X.T@temp_X)) @ temp_X.T@temp_y)
            y_hat2 = temp_X @ beta_hat
            syhat.append(y_hat2[-1])
        if plot:
            return syhat
        
        return np.mean(np.square(self.y[T0+h:]-syhat))
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], dayfirst = True)
def to_date_time(series, t0=0, h=0):
    test = pd.DataFrame(series)
    test = test.set_index(df['DATE'][t0+h:])
    test.index = pd.to_datetime(test.index, dayfirst=True)
    return test
```

```python
y = to_date_time(df['French_Retail_Sales'])
y = to_date_time(y)
plt.rcParams['figure.figsize'] = [15, 6]
def plot_graph(y_hat, t0, h, gdp=False):
    
    if gdp:
        yhat = to_date_time(y_hat, t0+1, h)
    else:
        yhat = to_date_time(y_hat, t0, h)
    
    plt.plot(yhat)
    plt.plot(y, alpha=0.4)
    return plt.show()
```

```python
#MSFE at h = 1
print(first_specification.MSFE_1(T0=50, h=1))
plot_graph(first_specification.MSFE_1(T0=50, h=1, plot=True),50,1)
```
