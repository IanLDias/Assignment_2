### Question 1

Specifications:


#### a) Describe the specifications

The dataset is of French retail sales. Four specifications are presented for how to estimate this data.

A note on why terms are included: The base year $t_0$ does not have a value of 0 and so the intercept $a_0$ is needed for all specificaitons. It is removed when it causes perfect multicollinearity, which happens when all dummies for the quarters or months are added (as in $S_2$ and $S_4$). The trend is included if the timeseries is thought to change with time, which looking at the graph above does and should be included in all specifications. Seasonality also looks to be present and so dummies for the quarter and month are included, with the last quarter and month being a time where spending is increased the most.

$S_1$: This includes an intercept - $a_0$, trend - $t$, and a dummy for the fourth quarter $D_4$

$S_2$: This includes a trend - $t$, and a dummy for all quarters $Di$, where $i$ $\epsilon$ [1,4]

$S_3$: This includes an intercept - $a_0$, a trend - $t$, and a dummy for the last month in the year $M$<sub>12</sub>

$S_4$: This includes a trend $t$, and a dummy for all months of the year - $M_i$, where $i$ $\epsilon$ [1,12]

For each specification, estimate the values of each parameter. Also compute MSE, AIC, BIC.

```python
class model:
    x = df['French_Real_GDP']
    
    def __init__(self, data, quarters=False, month=False, last=False, MSFE = False, GDP = False, h=0, april=False):
        ''' Quarters OR month'''
        self.data = data
        self.quarters = quarters
        self.month = month
        self.last = last 
        self.MSFE = MSFE
        self.GDP = GDP
        self.h = h
        self.x = self.x.shift(periods=h)[h:]
        self.y = df['French_Retail_Sales']
        self.y = self.y[h:]
        self.april = april
    def __repr__(self):

        return 'Quarters dummy: {}\nMonths dummy: {}\nOnly Last: {}\nParameters: {}\nModel Evals: {}'.format(
            self.quarters, self.month, self.last, self.parameter_estimation(), self.model_eval())
        #Print model specification, parameter estimation, model_eval

    def Specification(self):
        '''Changes the dataframe to be in quarter or month form'''
        df = pd.read_csv(self.data, sep = ';')
        df['DATE'] = pd.to_datetime(df['DATE'], dayfirst = True)
        if self.quarters:
            df['DATE']=df['DATE'].dt.quarter
            return df
        if self.month:
            df['DATE']=df['DATE'].dt.month
            return df
    
    def dummies(self):
        '''Makes dummies based on if quarter or month'''
        mod = self.Specification()
        X = pd.get_dummies(mod['DATE'])
        X = X.reset_index()
        X.rename(columns={'index':'t'}, inplace=True)
            #Also needs an intercept
        if self.last:
            X.insert(0,'a0',1)
            if self.GDP:
                X = X.join(self.x)
                X = X[self.h:]
                return X.iloc[:,:2].join(X.iloc[:,-2:])
            else:
                return X.iloc[:,:2].join(X.iloc[:,-1:])
        if self.GDP:
            X = X.join(self.x)
            X = X[self.h:]
        return X
    
    def parameter_estimation(self):
        '''Estimates the parameters for the model'''
        X = self.dummies()
        if self.GDP and self.last:
            #print(X)
            #print(self.y)
            a0, t, dummy, gdp = np.linalg.inv((X.T@X)) @ X.T@self.y
            return a0, t, dummy, gdp
        if self.last:
            a0, t, dummy = np.linalg.inv((X.T@X)) @ X.T@self.y
            return a0, t, dummy
        if self.GDP:
            #print(np.linalg.inv((X.T@X)) @ X.T@self.y)
            t, gdp, *dummy = np.linalg.inv((X.T@X)) @ X.T@self.y
            return t, gdp, dummy
        else:
            t, *dummy = np.linalg.inv((X.T@X)) @ X.T@self.y
            return np.append(t, dummy)
    
    def y_hat(self):
        if (self.GDP and not self.last):
            params = self.parameter_estimation()
            params1 = np.append(params[0], params[1])
            matrix = np.append(params1, params[-1])
            return self.dummies()@matrix.T
        matrix = np.array(self.parameter_estimation())
        return self.dummies()@matrix.T

    def model_eval(self):
        MSE = np.mean((self.y - self.y_hat())**2)
        T = len(self.y)
        AIC = T * MSE + (5*2)
        BIC = T * MSE + 5*np.log(T)
        return MSE, AIC, BIC
    
    def MSFE_1(self, T0, h, plot=False):
        X = self.dummies()
        T = len(self.y)
        syhat = []
        for i in range(T0, T+h):
            temp_y = self.y[:i]
            temp_X = X[:i]
            temp_X = np.array(temp_X)
            beta_hat = (np.linalg.inv((temp_X.T@temp_X)) @ temp_X.T@temp_y)
            y_hat2 = temp_X @ beta_hat
            syhat.append(y_hat2[-1])
        if plot:
            return syhat
        return np.mean(np.square(self.y[T0-h:]-syhat))
```

```python
y = df['French_Retail_Sales']
def var(Spec):
    k = len(Spec.dummies().columns)
    T = len(Spec.Specification())
    y_hat = Spec.y_hat()
    return np.sum(((y-y_hat)**2))*(1/(T-k)), y_hat
```

```python
#Simple exponential smoothing (SES) - no clear trend or seasonality
#Alpha is the rate at which the weights decrease (% of current value) Large alpha (close to 1) gives less weight to older values
#alpha = level, beta = trend, gamma = seeasonality. Parameters smooths each = 0.4. 
#m = frequency of the seasonality (4 or 12) in a year
#Seasonal = addititve? 
```

```python
y = to_date_time(df['French_Retail_Sales'])
y = y['French_Retail_Sales']
```

```python
def HWS(T0, s, alpha, beta, gamma, h, T = len(y)):
    syhat = []
    S = np.zeros((1,T-h))[0]
    L = np.mean(y[1:s])
    b = 0
    S[1:s] = y[1:s] - L
    
    for t in range(s+1, T-h):
        newL = alpha * (y[t] - S[t-s]) + (1-alpha) * (L+b)
        newb = beta * (newL-L) + (1-beta)*b

        S[t] = gamma * (y[t] - newL) + (1-gamma)*S[t-s]
        yhat = newL + h*newb + S[t+h-s]
        L = newL
        b = newb
        if t >= T0:
            syhat.append(yhat)
    return syhat
```
